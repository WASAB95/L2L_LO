```python
neurons = 10

class Network(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Inputs to hidden layer linear transformation
        self.fa = nn.Linear(1, neurons)
        self.fao = nn.Linear(neurons, 1)
        # Output layer
        self.fp = nn.Linear(1, neurons)
        self.fpo = nn.Linear(neurons, 1)
        
        # Define reulu activation and softmax output 
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        amp_x = self.fa(x)
        amp_x = self.relu(amp_x)
        amp = self.fao(amp_x)
        ph_x = self.fp(x)
        ph_x = self.relu(ph_x)
        ph = self.fpo(ph_x)
        
        return {'amp': amp, 'phase': ph}
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
loss_list = []
steps = 0
for epoch in range(num_epochs):
    for idx, (data, targets) in enumerate(train_loader):
        
        data = data.view(data.size(0),1).to(device)
        amp_target = targets[0][0].view(1).to(device)
        ph_target = targets[0][1].view(1).to(device)
        
#         print(data)
        
#         print(data.size)
#         print(amp_target)
#         print(ph_target)
        
        optimizer.zero_grad()
        
        
        output = model(data[0])
        amp_out = output['amp'][0]
        phase_out = output['phase'][0]
        
#         print("target")
#         print(amp_target)
#         print("output")
#         print(amp_out)

        
        
        loss = loss_fn(amp_out, amp_target)
        loss = loss + loss_fn(phase_out, ph_target)
        loss_list.append(loss.item())
        
        
        
        loss.backward()
        optimizer.step()
#     print(idx)
        
    print(f'epoch {epoch+1} / {num_epochs}, setp {idx+1}, loss={loss.item()}')
    steps += idx+1
#     print(steps)
```

```python
from collections import namedtuple
from typing import Any

import torch


# pylint: disable = abstract-method
class ModelWrapper(torch.nn.Module):
    """
    Wrapper class for model with dict/list rvalues.
    """

    def __init__(self, model: torch.nn.Module) -> None:
        """
        Init call.
        """
        super().__init__()
        self.model = model

    def forward(self, input_x: torch.Tensor) -> Any:
        """
        Wrap forward call.
        """
        data = self.model(input_x)

        if isinstance(data, dict):
            data_named_tuple = namedtuple("ModelEndpoints", sorted(data.keys()))  # type: ignore
            data = data_named_tuple(**data)  # type: ignore

        elif isinstance(data, list):
            data = tuple(data)

        return data
```
